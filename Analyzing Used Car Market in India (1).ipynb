# 2>.astype(str): -Convert all values in the Mileage column to strings.
### Why this accurs
    - Because .str.extract() only works on strings.
    - Example: '20.5 kmpl', '22.4 km/kg'.

# 3>.str.extract(r'(\d+\.\d+|\d+)'):
    - Use a regular expression (regex) to extract only the numeric part from the string.
    - \d+\.\d+ matches a decimal number (e.g., 20.5)
    - \d+ matches an integer (e.g., 20)
    - The whole pattern means: “extract either a decimal number or an integer”

### Examples:1>'20.4 kmpl' → 20.4  2>'23 km/kg' → 23

# 4>.astype(float):
    - Convert the extracted string number into a float type so it can be used in numerical operations or visualizations.

```python
# This Code Step shown above
df['Mileage'] = df['Mileage'].astype(str).str.extract(r'(\d+\.\d+|\d+)').astype(float)
df['Engine'] = df['Engine'].astype(str).str.extract(r'(\d+)').astype(float)
df['Power'] = df['Power'].astype(str).str.extract(r'(\d+\.\d+|\d+)').astype(float)
```

# D. Outlier Detection
- Outlier detection is the process of identifying data points that are significantly different from the majority of the dataset. These unusual values can distort statistical analyses and machine learning models, so they are an important part of EDA.

- Why Detect Outliers?
    - They can skew your mean, standard deviation, regression lines, etc.
    - They might indicate data entry errors or exceptional cases.
    - In some cases, outliers contain valuable insights (e.g., fraud detection).
- ex1> sns.boxplot(df['sales'])

- EX2>Q1 = df['Price'].quantile(0.25)
- Q3 = df['Price'].quantile(0.75)
- IQR = Q3 - Q1
- outliers = df[(df['Price'] < Q1 - 1.5 * IQR) | (df['Price'] > Q3 + 1.5 * IQR)]


# E. Feature Distribution
- Feature distribution refers to how the values of a particular column (feature) are spread across the dataset. It is one of the most fundamental steps in EDA because it helps you understand the nature of your data — whether it's normal, skewed, categorical, or has outliers.

-  Why is Feature Distribution Important?
    -  Helps identify data types (categorical vs numerical)
    - Reveals data imbalance (e.g., more petrol cars than diesel)
    - Shows skewness, outliers, multimodal behavior
    - Aids in deciding transformations or scaling
    - Guides feature selection and model building

- ex1>df.hist(figsize=(10, 8))
